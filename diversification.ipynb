```python
import hvplot.pandas
import yfinance as yf
from yahoo_fin.stock_info import *
%store -r tickers_list
import numpy as np

data = yf.download(tickers_list, period='36mo')['Adj Close']

#Finding daily returns of all the stocks
daily_returns = data.pct_change().dropna()

#Calculating variance of the s&p 500
print("Variance of the s&p 500")

SP_variance=daily_returns["SPY"].var()
display(SP_variance)


#Calculating Covariance and beta, beta avg of  chosen stocks using 60 day rolling window

index = 0
while index < len(tickers_list)-1:
    stock = tickers_list[index]
    stockcovariance =  daily_returns.rolling(60).cov(daily_returns["SPY"])
    stockcovariance.dropna().tail()
    
    print(f"The beta of {stock}")
    stockbeta = stockcovariance[stock] / SP_variance
    stockbeta.dropna().tail()
   
    print(f"{stock} beta average is")
    stockbeta_average = stockbeta.mean()
    print(stockbeta_average)
    display(stockbeta.hvplot(title="60 days rolling stock beta", group_label='Tickers'))
    index += 1

stock.plot(df, type='candle', style='charles',
            title='stock',
            ylabel='Price',
            ylabel_lower='Volume',
            volume=True,
            mav=(1,3,6))    
```
