```python
import hvplot.pandas
import yfinance as yf
from yahoo_fin.stock_info import *
%store -r tickers_list
import numpy as np
import csv

data = yf.download(tickers_list, period='36mo')['Adj Close']

#Finding daily returns of all the stocks
daily_returns = data.pct_change().dropna()

#Calculating standard deviation
standard_deviation = daily_returns.std().sort_values()
display(f"Your standard deviations",standard_deviation)

#calculating annualized standard deviation.
annualized_standard_deviation = standard_deviation * np.sqrt(252)
annualized_standard_deviation

#using 21 day rolling window to plot standard deviation
rolling_deviation21 = daily_returns.rolling(window=21).std().hvplot(title="21 day rolling deviation", group_label='Tickers')
display(rolling_deviation21)

#Calculating average annual return for each of the Stocks.
year_trading_days=252
average_annual_return_portfolio = daily_returns.mean() * year_trading_days
average_annual_return_portfolio.sort_values()

#Finding Sharpe ratios for all 5 stocks and S&P
sharpe_ratios = average_annual_return_portfolio / standard_deviation
sharpe_ratios.sort_values()

#Plotting Sharpe ratios
display(sharpe_ratios.hvplot.bar(
    height=500,
    width= 700,
    title="Sharpe Ratio Of Tickers"))

#Calculating variance of the s&p 500
print("Variance of the s&p 500")

SP_variance=daily_returns["SPY"].var()
display(SP_variance)


#Calculating Covariance of 5 chosen stocks using 60 day rolling window
stockcovariance =  daily_returns.rolling(60).cov(daily_returns["SPY"])
stockcovariance.dropna().tail()
display(stockcovariance.hvplot(
        height = 300,
        width = 700,
        title = "Covariance Of Chosen Stocks"))

```
