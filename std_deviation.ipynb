```python
import hvplot.pandas
import yfinance as yf
from yahoo_fin.stock_info import *
%store -r tickers_list
import numpy as np

data = yf.download(tickers_list, period='36mo')['Adj Close']

#Finding daily returns of all the stocks
daily_returns = data.pct_change().dropna()

#Calculating standard deviation
standard_deviation = daily_returns.std().sort_values()
display(f"Your standard deviations",standard_deviation)

#calculating annualized standard deviation.
annualized_standard_deviation = standard_deviation * np.sqrt(252)
annualized_standard_deviation

#using 21 day rolling window to plot standard deviation
rolling_deviation21 = daily_returns.rolling(window=21).std().hvplot(title="21 day rolling deviation", group_label='Tickers')
display(rolling_deviation21)

#Calculating average annual return for each of the Stocks.
year_trading_days=252
average_annual_return_portfolio = daily_returns.mean() * year_trading_days
average_annual_return_portfolio.sort_values()

#Finding Sharpe ratios for all 5 stocks and S&P
sharpe_ratios = average_annual_return_portfolio / standard_deviation
sharpe_ratios.sort_values()

#Plotting Sharpe ratios
display(sharpe_ratios.hvplot.bar(
    height=500,
    width= 400,
    title="Sharpe Ratio Of Tickers"))

#Calculating variance of the s&p 500
print("Variance of the s&p 500")

SP_variance=daily_returns["SPY"].var()
display(SP_variance)


#Calculating Covariance and beta, beta avg of  chosen stocks using 60 day rolling window

index = 0
while index < len(tickers_list)-1:
    stock = tickers_list[index]
    stockcovariance =  daily_returns.rolling(60).cov(daily_returns[SPY])
    stockcovariance.dropna().tail()
    print(f"The beta of {stock}")
    stockbeta = stockcovariance[stock] / SP_variance
    display(stockbeta.dropna().tail())
    print(f"{stock} beta average is")
    stockbeta_average = stockbeta.mean()
    print(stockbeta_average)
    index += 1

#stock1,stock2,stock3,stock4,stock5,stock6 = tickers_list
 #Calculating the beta of each of the stocks.

stockcovariance =  daily_returns.rolling(60).cov(daily_returns["SPY"])
stockcovariance.dropna().tail()

index = 0
while index < len(tickers_list)-1:
    stock = tickers_list[index]
    stockbeta = stockcovariance[stock] / SP_variance
    print(stockbeta)
    index += 1





print(f"The beta of {stock1}")
stockbeta1 = stockcovariance[stock1] / SP_variance
display(stockbeta1.dropna().tail())



print(f"The beta of {stock2}")
stockbeta2 = stockcovariance[stock2] / SP_variance
display(stockbeta2.dropna().tail())

print(f"The beta of {stock3}")
stockbeta3 = stockcovariance[stock3] / SP_variance
display(stockbeta3.dropna().tail())

print(f"The beta of {stock4}")
stockbeta4 = stockcovariance[stock4] / SP_variance
display(stockbeta4.dropna().tail())

print(f"The beta of {stock5}")
stockbeta5 = stockcovariance[stock5] / SP_variance
display(stockbeta5.dropna().tail())

 #Caculating the beta average for each of the 5 stocks
print(f"{stock1} beta average is")
stockbeta1_average = stockbeta1.mean()
print(stockbeta1_average)

print(f"{stock2} beta average is")
stockbeta2_average = stockbeta2.mean()
print(stockbeta2_average)

print(f"{stock3} beta average is")
stockbeta3_average = stockbeta3.mean()
print(stockbeta3_average)

print(f"{stock4} beta average is")
stockbeta4_average = stockbeta4.mean()
print(stockbeta4_average)

print(f"{stock5} beta average is")
stockbeta5_average = stockbeta5.mean()
print(stockbeta5_average)

#  #plotting the beta of every stock using matplotlib
# tikers_list = [stock1    
# fig, ax = plt.subplots(figsize=(15, 5))

# ax.plot(stockbeta1, color = 'green', label = f"{stock1}")
# ax.plot(stockbeta2, color = 'red', label = f"{stock2}")
# ax.plot(stockbeta3, color = 'blue', label = f"{stock3}")
# ax.plot(stockbeta4, color = 'yellow', label = f"{stock4}")
# ax.plot(stockbeta5, color = 'purple', label = f"{stock5}")
# ax.set_title("Stock betas over 1 year of trading")
# ax.legend(loc = 'upper left')
# plt.show()
```
