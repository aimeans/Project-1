```python
import hvplot.pandas
import yfinance as yf
from yahoo_fin.stock_info import *
%store -r tickers_list

data = yf.download(tickers_list, period='36mo')['Adj Close']
    
#data.to_csv('ticker.csv')

#Finding daily returns of all the stocks
daily_returns = data.pct_change().dropna()

## Measuring Volatility

display(daily_returns.hvplot(
    title="daily returns over 3 years",
    ylabel='Daily Returns',
    xlabel='Date',
    group_label='Tickers'
    #height=700
 
))



#Finding cumulative returns and displaying last 5 rows
cumulative_returns =(1 + daily_returns).cumprod()
#cumulative_returns.tail()

#Plotting Cumulative returns
display(cumulative_returns.hvplot(
    title="Cumulative Returns over 3 years",
    ylabel='Cumulative Returns',
    xlabel='Date',
    group_label='Tickers'
 
))

%store daily_returns
%store cumulative_returns
```
