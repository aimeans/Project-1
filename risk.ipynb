## Risk Analysis

```python
import hvplot.pandas
import yfinance as yf
from yahoo_fin.stock_info import *
%store -r tickers_list
import numpy as np

data = yf.download(tickers_list, period='36mo')['Adj Close']

#Finding daily returns of all the stocks
daily_returns = data.pct_change().dropna()

#Calculating standard deviation
standard_deviation = daily_returns.std().sort_values()
#display(f"Your standard deviations",standard_deviation)

# Calculate and sort the annualized standard deviation (252 trading days) of the 4 portfolios and the S&P 500
# Review the annual standard deviations smallest to largest
trading_days = 252

annualized_standard_deviation = standard_deviation * trading_days**(1/2)

# Display the sorted results
annualized_standard_deviation.sort_values()

#using 30 day rolling window to plot standard deviation
rolling_std_30 = daily_returns.rolling(window=30).std().hvplot(title="30 day rolling deviation", group_label='Tickers')
display(rolling_std_30)

#using 30 day rolling window to plot standard deviation
stocks_daily_returns = daily_returns.drop(['SPY'], axis='columns')
stocks_rolling_deviation_30 = stocks_daily_returns.rolling(window=30).std().hvplot(title="30 day rolling deviation", group_label='Tickers')
display(stocks_rolling_deviation_30)

#using 180 day rolling window to plot standard deviation
rolling_std_180 = daily_returns.rolling(window=180).std().hvplot(title="180 day rolling deviation", group_label='Tickers')
display(rolling_std_180)

#using 180 day rolling window to plot standard deviation
stocks_daily_returns = daily_returns.drop(['SPY'], axis='columns')
stocks_rolling_deviation_180 = stocks_daily_returns.rolling(window=180).std().hvplot(title="180 day rolling deviation", group_label='Tickers')
display(stocks_rolling_deviation_180)



#%store standard_deviation
%store stocks_daily_returns
%store stocks_rolling_deviation_30
%store stocks_rolling_deviation_180
```
